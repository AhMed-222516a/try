```python
def factorial(num):
    if num==0:
        return 1
    elif num==1:
        return 1
    else:
        return num*factorial(num-1)
factorial(5)
    
```

```python
def isPrime(num):
    for i in range (2,num):
        if num%i==0:
            return False
    return True
isPrime(20)
```

```python
def dividors(num1,num2):
    """
    param_1 : take first number from user
    type:pram_1: int    
    param_2 : take scond number from user
    type:pram_1: int
    return: list of common dividors
    typer_return:list
    """
    mn=min(num1,num2)
    common_dividors=[]
    for i in range(1,mn+1):
        if num1%i==0 and num2%i==0:
            common_dividors.append(i)
    return common_dividors
dividors(20,30)
```
