```python
class FibonacciTask:
  """
  task_1 class calcolate the fibonacchi for length num
  the therd number in the list is the sum of the brevios two number
  """
  def __init__(self,num) :
    self.num=num
  def fib (self):
    """
    pram_1: take the number of output for fibonacchi
    type pram_1: Int

    return: a list of fibonacchi with size of num
    type return: list
    """
    if self.num == 1:return [0]
    if self.num <= 0:return "error"
    out=[0,1]
    for i in range(self.num-2):
      out.append((out[i+1]+out[i]))
    return out
f = FibonacciTask(10)
print(f.fib())
f.num=5
print(f.fib())
```

```python
class remove_an_item:
  """
  ItemRemover class handles removing a specific item from a set.
  Attributes:
     set_ (set): The original set.
     remove_item (any): The item to be removed.
  """

  def __init__(self,set_,remove_Item):
    self.set_=set_
    self.remove_Item=remove_Item

  def rem(self):
    """
    pram_1: take the set from user to remove item from it
    type pram_1: set

    return: a set that heve remove item from it
    type return: set
    """
    b = list(self.set_)
    b.remove(self.remove_Item)
    return set(b)

p=remove_an_item({1, 2, 3, 4, 5},4)
print(p.rem())
```

```python
class find_the_index:
  """
  task 3 he find the index from the tuple
  Attrebutes:
    my_tuple (tuple) : the original tuple
    index_   (any)   : the string you serch it
  """

  def __init__(self,my_tuple,index_):
    self.my_tuple=my_tuple
    self.index_=index_

  def index_tuple(self):
    """
    pram_1: take the tuple from user
    type pram_1: tuple
    pram_2: take the item you want to find it
    type pram_2: tuple or string
    """
    x= [b for b in range(len(self.my_tuple)) if self.my_tuple[b] == self.index_]
    return x

f=find_the_index(('apple', 'banana', 'orange', 'grape', 'banana'),'orange')
print(f.index_tuple())
```

```python
class sentence_into_list:
  """
  task 4 he convert the string to list
  Attrebutes:
    word (string) : a string of data separated by space
  """

  def __init__ (self,word):
    self.word=word

  def list_of_words(self):
    """
    pram_1: input the word you want to split according to space only
    type pram_1: string

    return: a list of words
    type return: list
    """
    #to check if the input is string or not
    if not isinstance(self.word,str):return ("you shoud enter string")
    return self.word.split()

print(sentence_into_list("Welcome to Amit").list_of_words())
```

```python
class Palindrome_print:
  """
  task 5 he return if the word is Palindrome or not
  Attrebutes:
    word (string) : a word string
  """

  def __init__ (self,word):
    self.word=word

  def palindrome (self):
    """
    pram_1: he take a word you want to check if it is palindrome
    type pram_1: string

    retunr: a word that tell you if it palindrome or not palindrome
    tyoe return: string
    """
    #to check if the input is string or not
    if not isinstance(self.word,str):return ("you shoud enter string")

    if len(self.word)%2 == 0 :return "not palindrome"
    for i in range(int(len(self.word)/2)):
      if self.word[i] != self.word[-i-1]: return "not palindrome"
      else: return "palindrome"

print(Palindrome_print("racecar").palindrome())
```

```python
class find_the_factorial:
  """
  task 6 he return the factorial of number
  Attrebutes:
    num (int) : integar number
  """

  def __init__(self,num):
    self.num=num

  def factorial(self):
    """
    pram_1: is the number you want to git his factorial
    type pram_1: int

    return: the factorial
    type:int
    """
    if not isinstance(self.num,int): return "error" #check if it is integar or not
    if self.num<0: return "error" # check if it is positive or not
    if self.num == 0 or self.num ==1 : return 1 #check if it is 0 or 1 case it is factorial= 1
    result=1
    for i in range(self.num):
      result*=(i+1)
    return result

print(find_the_factorial(5).factorial())
```

```python
class List_of_Squares:
  """
  task 7 he a list of square numbers from 1 to num you enter
  Attrebutes:
    num (int) : integar number
  """

  def __init__(self,num):
    self.num=num

  def squares_num(self):
    """
    pram_1: is the number in the list you want see her squares form zero to --> num
    type pram_1: int

    return: list of square number from
    type return: list
    """
    if not isinstance(self.num,int): return "error"
    if self.num < 1: return "error"
    out=[]
    for i in range (self.num):
      out.append((i+1)**2)
    return out

  def squares_num_another_solution(self):
    """
    pram_1: is the number in the list you want see her squares form zero to --> num
    type pram_1: int

    return: list of square number from
    type return: list
    """
    if not isinstance(self.num,int) or self.num < 1: return "error"
    return [i**2 for i in range(1,self.num+1)]

print(List_of_Squares(10).squares_num())
```

```python
class Sum_of_Even_Numbers:
  """
  task 8 he calculate the sum of list of numbers
  Attrebutes:
    num (list) : list of number
  """

  def __init__(self,num):
    self.num=num

  def even_sum(self):
    """
    pram_1: take the list of data you want sum the even number of it
    type pram_1: list

    return: the sum of all even in num
    type return: int
    """
    if not isinstance(self.num,list): return "error"
    self.num.sort()
    sum=0
    for i in range(len(self.num)):
      if (self.num[i]%2)==0: sum+=self.num[i]
    return sum

  def even_sum_another_solution(num_list):
    """
    pram_1: take the list of data you want sum the even number of it
    type pram_1: list

    return: the sum of all even in num
    type return: int
    """
    if not isinstance(num_list,list): return "error"
    return sum(i for i in num_list if (i%2) ==0)


print(Sum_of_Even_Numbers([1, 2, 3, 4, 5, 6]).even_sum())
```

```python
class median_of_list :
  """
  task 10 he return the median of data
  Attrebutes:
    list_data (list) : list of numbers
  """

  def __init__(self,list_data):
    self.list_data=list_data

  def median(self):
    """
    pram_1: is a list of data ypu want to gir his median
    type pram_1:list

    return: the median number of list_data
    type return:int
    """
    if not isinstance(self.list_data,list):return "error"
    self.list_data.sort()
    return self.list_data[int(len(self.list_data)/2)]
print(median_of_list([5,2,7,3,1]).median())
```

```python
class convert_to_grade:
  def __init__ (self,degree):
    self.degree=degree
  def grades(self):
    if not isinstance(self.degree,int): return None
    if 100>= self.degree >= 85: return "A"
    if 85 >  self.degree >= 75 : return "B"
    if 75 >  self.degree >= 65 : return "C"
    if 65 >  self.degree >= 50 : return "D"
    if 50 >  self.degree      : return "F"
print(convert_to_grade(100).grades())
```
