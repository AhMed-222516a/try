Part 1: Smart Array Factory
Task: Create a function that allows students to generate specific array types quickly using built-in NumPy methods.


```python
import numpy as np
import difflib

def normalize_mode(mode):
    mode = mode.lower()
    VALID_MODES = ['zeros', 'ones', 'full', 'identity']
    match = difflib.get_close_matches(mode, VALID_MODES, n=1, cutoff=0.6)
    if match:
        return match[0] 
    else:
        return mode  

def array_factory(mode, shape, value=None):
    """
    Creates various NumPy arrays based on the mode.
    - 'zeros': Array filled with 0.
    - 'ones': Array filled with 1.
    - 'full': Array filled with a specified 'value'.
    - 'identity': A square identity matrix of size 'shape'.
    """
    if not isinstance(mode,str) : 
        raise ValueError("mode should be string")    
    mode = mode.lower()
    mode = normalize_mode(mode)
    if mode == 'zeros':
        return np.zeros(shape)

    elif mode == 'ones':
        return np.ones(shape)

    elif mode == 'full':
        if value is None:
            raise ValueError("You must provide a value for 'full' mode")
        return np.full(shape, value)

    elif mode == 'identity':
        if not isinstance(shape, int):
            raise ValueError("Shape must be an integer for identity matrix")
        return np.identity(shape)

array_factory('zero',(3,3))
```
