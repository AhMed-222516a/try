```python
#add to lsts not shoud same order
def list_add(lst1, lst2):
    """
    pram_1: take the first list from user
    type pram_1: list
    pram_2: take the second list from user
    type pram_2: list

    return: return a new suming list
    trupe retrun: list
    """
    new=[]
    small=min(len(lst1),len(lst2))
    big=max(len(lst1),len(lst2))
    for i in range(big):
        if i<small:
            new.append(lst1[i]+lst2[i])
        else : 
            new.append(max(lst1,lst2,key=len)[i])
        
    return new
print(list_add([1,2,3,4,10],[4,5,6]))
```

```python
# I 1 V 5 X 10 L 50 C 100 D 500 M 1000
def roman_to_int(s):
    """
    pram_1: take the roman number from user
    type pram_1: str

    return: return a new suming list
    trupe retrun: int
    """
    roman_dict = {'I': 1, 'V': 5, 'X': 10, 'L': 50, 'C': 100, 'D': 500, 'M': 1000}
    total = 0
    prev_value = 0
    
    for char in s:
        value = roman_dict[char]
        if prev_value < value:
            total += value - 2 * prev_value
        else:
            total += value
        prev_value = value      
    return total
print(roman_to_int("MCMXCIV"))
```

```python
def searchInsert(nums, target):
    """
    pram_1: take the sorted list from user
    type pram_1: list
    pram_2: take the target value from user
    type pram_2: int

    return: return the index of the target value if found, otherwise return the index where it would be if it were inserted in order
    trupe retrun: int
    """
    if not isinstance(nums, list) or not isinstance(target, int):
        return "error"
    left = 0
    right = len(nums) - 1

    while left <= right:
        mid = (left + right) // 2

        if nums[mid] == target:
            return mid
        elif nums[mid] < target:
            left = mid + 1
        else:
            right = mid - 1

    return left
searchInsert([1,3,5,6], 5)
```

```python
def calculate(s: str) -> int:
    stack = []
    num = 0
    sign = '+'
    
    s += '+'   
    for ch in s:
        if ch.isdigit():
            num = num * 10 + int(ch)
        
        elif ch in '+-*/':
            
            if sign == '+':
                stack.append(num)
            
            elif sign == '-':
                stack.append(-num)
            
            elif sign == '*':
                stack.append(stack.pop() * num)
            
            elif sign == '/':
                last = stack.pop()
                stack.append(int(last / num))  # truncate toward zero
            
            sign = ch
            num = 0
    
    return sum(stack)
print(calculate("3+2*2"))
```
