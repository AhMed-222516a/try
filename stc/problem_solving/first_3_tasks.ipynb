```python
class who_play_banjo:
  """
  this function takes a string and 
  check if the first letter is r or R then return that the person plays banjo 
  otherwise return that the person does not play banjo
  attributes:
    name: str
  
  
  """
  def __init__ (self,name):
    self.name=name
  def __str__(self) :
    return self.name
  def banjo(self):
    if self.name.lower().strip()[0]=='r':
        return self.name + " plays banjo"
    else:
      return self.name + " does not play banjo"
guest=who_play_banjo('Ramey')
print(guest.banjo())    
      
```

```python
class Big:
    def __init__(self, num, target):
        if not isinstance(num, list):
            raise TypeError("num must be a list")

        if not isinstance(target, int):
            raise TypeError("target must be an integer")

        self.num = num
        self.target = target

    def two_sum(self):
        seen = {}

        for i, value in enumerate(self.num):
            complement = self.target - value

            if complement in seen:
                return f"{complement} + {value}"

            seen[value] = i

        return "No solution found"
    
num , target = [11,2,4,7,8,15] , 9
p=big(num,target)
p.two_sum()
```
