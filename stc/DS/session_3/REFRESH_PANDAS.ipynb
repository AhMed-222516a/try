```python
import numpy as np

# 1D array with 6 elements
arr = np.array([1, 2, 3, 4, 5, 6])

print("Original array:")
print(arr)
print("Original shape:", arr.shape)

# Reshape to 2D (3 rows, 2 columns)
new_arr = arr.reshape(2, )

print("\nReshaped array:")
print(new_arr)
print("New shape:", new_arr.shape)
```
