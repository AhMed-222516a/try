```python
class word_chane:
  """
  this function takes a string and reverse the order of the words in it
  attributes:
    word: str
  """
  def __init__(self,word):
    self.word=word
  def __str__(self):
    return self.word
  def reverse_phrases(self):
    return str(self.word.split()[::-1])

p=word_chane("hello world")
print(p)
```

```python
class who_play_banjo:
  """
  
  
  """
  def __init__ (self,name):
    self.name=name
  def __str__(self) :
    return self.name
  def banjo(self):
    if self.name.lower().strip()[0]=='r':
        return self.name + " plays banjo"
    else:
      return self.name + " does not play banjo"
guest=who_play_banjo('Ramey')
print(guest.banjo())    
      
```
