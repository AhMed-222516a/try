```python
import numpy as np
import matplotlib.pyplot as plt
x= np.array([1,2,3,4])
y= np.array([2,2.2,3.6,4.7])

w,b,alpha=0,0,0.01
num_iteration=20

sse_values = []
for i in range(num_iteration):
    y_hat = w*x + b
   
    D_w = 2 * np.sum((y_hat - y)*x)
    D_b = 2 * np.sum(y_hat - y)
   
    w -= alpha * D_w
    b -= alpha * D_b
   
    sse = (np.sum((y_hat - y )**2))
    sse_values.append(sse)
   
    if (i+1) % 20 == 0:
        print(f"interation {i+1},SSE: {sse}")
 
plt.figure(figsize=(12,5))
 
plt.subplot(1,2,1)
plt.plot(range(num_iteration), sse_values, label = 'SSE')
plt.xlabel('Iteration')
plt.ylabel('SSE')
plt.title('SSE oner iteration')
plt.legend()
 
plt.subplot(1,2,2)
 
plt.scatter(x,y, color = 'blue', label ='Data points')
plt.plot(x, w*x + b , color = 'red', label = 'Regression line' )
plt.show()
```
