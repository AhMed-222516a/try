```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from sklearn.model_selection import train_test_split # for splitting the dataset into training and testing sets
from sklearn.linear_model import LinearRegression #for building the linear regression model
from sklearn.metrics import mean_squared_error, r2_score,root_mean_squared_error #for valuating the performance of the model
import warnings
warnings.filterwarnings('ignore') #to ignore any warnings that may arise during the execution of the code
```

```python
data =pd.read_csv('Salary_Data.csv')
```

```python
data.head()
```

```python
plt.figure(figsize=(3,3))
sns.pairplot(data,x_vars=['YearsExperience'], y_vars=['Salary'], size=7,kind='scatter')
plt.xlabel('Years')
plt.ylabel('Salary')
plt.title('Salary prediction')
plt.show()
```

```python
x = data.iloc[:,:-1] # selecting all rows and all columns except the last one as features
y = data.iloc[:,1] # selecting all rows and the second column (index 1) as the target variable
x_train, x_test, y_train, y_test = train_test_split(x , y , train_size = 0.8 , random_state = 10 )# splitting the dataset into training and testing sets with 80% for training and 20% for testing, and setting a random state for reproducibility
```

sklean ده معنا ال box 

```python
linear_model = LinearRegression() # creating an instance of the LinearRegression class
linear_model.fit(x_train,y_train) # fitting the linear regression model to the training data
```

```python
plt.scatter(x_train,y_train,color='red') # plotting the training data points in red
plt.plot(x_train,linear_model.predict(x_train),color='blue') # plotting the regression line in blue
plt.xlabel('Years of Experience') # setting the x-axis label
plt.ylabel('Salary') # setting the y-axis label
plt.title('salary vs Experience(training set)') # setting the title of the plot
plt.show() # displaying the plot
```

```python
plt.scatter(x_test,y_test,color='red') # plotting the testing data points in red
plt.plot(x_train,linear_model.predict(x_train),color='blue') # plotting the regression line in blue
plt.xlabel('Years of Experience') # setting the x-axis label
plt.ylabel('Salary') # setting the y-axis label
plt.title('salary vs Experience(test set)') # setting the title of the plot
plt.show() # displaying the plot
```

```python
y_test
```

```python
y_pred = linear_model.predict(x_test) # making predictions on the testing set using the fitted model
y_pred
```

```python
mse = mean_squared_error(y_test,y_pred) # calculating the mean squared error between the actual target values and the predicted values
print(mse)
rmse =root_mean_squared_error(y_test,y_pred) # calculating the root mean squared error between the actual target values and the predicted values
print(rmse) # the diffrence between y predict and test in the range of rmse 
```
