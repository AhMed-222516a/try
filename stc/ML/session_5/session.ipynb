```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('Salary_Data_v2.csv')
data.head()
```

```python
x = data.iloc[:,:-1]
y = data.iloc[:,1]
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.8,random_state=10)
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(degree=15) # make dgree then 15 and see the graph
x_poly_train = poly.fit_transform(x_train)
x_poly_test = poly.fit_transform(x_test)

# print(x_poly_train)
poly_model = LinearRegression()
poly_model.fit(x_poly_train,y_train)
```

```python
y_poly_pred_train = poly_model.predict(x_poly_train)
y_poly_pred_test = poly_model.predict(x_poly_test)
```

```python
poly_model = LinearRegression()
poly_model.fit(x_poly_train, y_train)
y_poly_pred_train = poly_model.predict(x_poly_train)
y_poly_pred_test = poly_model.predict(x_poly_test)
```

```python
x_range = np.linspace(x.min(), x.max(), 100).reshape(-1, 1)
x_range_poly = poly.transform(x_range)
y_range_pred = poly_model.predict(x_range_poly)
```

```python
plt.scatter(x_train,y_train,color='red',label='Training Data')
plt.scatter(x_test,y_test,color='Green',label='Testing Data')
plt.plot(x_range,y_range_pred,color='blue',label='Polynomial Regression Line')
plt.title("Salary vs Experience")
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.legend()
plt.show()
```
