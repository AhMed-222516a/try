```python
import numpy as np
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
```

```python
x = np.array([[1], [2], [3], [4]])
y = np.array([2, 3, 5, 7])
```

```python
lin_reg = LinearRegression()
lin_reg.fit(x, y)
y_pred_lin = lin_reg.predict(x)

lin_reg.coef_
lin_reg.intercept_
```

```python
ridge = Ridge(alpha = 50)# 10 40 50 for alpha
ridge.fit(x, y)
y_pred_ridge = ridge.predict(x)


ridge.coef_
ridge.intercept_
```

```python
lasso = Lasso(alpha = 10) 
lasso.fit(x, y)
y_pred_lasso = lasso.predict(x)

lasso.coef_
lasso.intercept_
```

```python
print("Linear Regression \nMSE:",mean_squared_error(y, y_pred_lin))
print("R2:",r2_score(y, y_pred_lasso))
print("="*50)
print("Linear Regression MSE:", mean_squared_error(y, y_pred_lin))
print("Ridge Regression MSE:", mean_squared_error(y, y_pred_ridge))
print("Lasso Regression MSE:", mean_squared_error(y, y_pred_lasso))
```

```python
plt.figure(figsize=(10, 6))
plt.subplot(1,3,1)
plt.scatter(x,y,color='blue',label='Actual Data')
plt.plot(x, y_pred_lin, color='red', label='Linear Regression')
plt.title('Linear Regression')
plt.legend()
```

```python
plt.figure(figsize=(10, 6))
plt.subplot(1,3,1)
plt.scatter(x,y,color='blue',label='Actual Data')
plt.plot(x, y_pred_ridge, color='red', label='Ridge Regression')
plt.title('Ridge Regression')
plt.legend()
```
